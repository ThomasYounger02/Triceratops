```python
import pandas as pd
import numpy as np
```

# 创建

## 传入等长列表/数组组成的字典

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
    'year':[2000, 2001, 2002, 2001, 2002],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9]
}

df1 = pd.DataFrame(
    data, 
    columns=['year', 'state', 'pop'],               # 可以指定列的顺序
    index=['one', 'two', 'three', 'four', 'five']   # 自定义index
)

df1
```

## 嵌套字典

```python
# 嵌套字典
pop = {
    'Nevada':{               # 外层字典-键：列
        2001:2.4, 2002:2.9   # 内层字典-键：index
    },
    'Ohio':{
        2000:1.5, 2001:1.7, 2002:3.6   # 不用等长
    }
}
df2 = pd.DataFrame(pop)
df2
```

# 属性

```python
df2.index.name = 'year'
df2.columns.name = 'state'

df2
```

```python
df2.values         # values时numpy数组
```

## 赋值-另一个Series
将列表/数组赋值给某个列，其长度必须跟DF的长度匹配。
如果赋值的是一个Series，会精确匹配DF的索引，所有的空位都会填上缺失值

```python
df1
```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
df1['debt'] = val   # 按索引匹配，缺失值为空

df1
```

## df和series之间的运算-沿着行广播

```python
df = pd.DataFrame(np.arange(12.).reshape(4, 3), columns=list('bde'),
    index=['Utah', 'Ohio', 'Texas', 'Oregon'])
s1 = df.iloc[0]      # 第一行
s2 = pd.Series(range(3), index=list('bef'))

df
```

```python
df - s1   # 将Seriesd的索引匹配到DF的列，并沿着行向下广播
```

```python
df + s2  # 不重叠的列为NaN
```

## 沿着列广播

```python
s3 = df['d']          # 第二列

df.sub(s3, axis=0)    # 每一列分别和s3相减
```

# 合并

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', np.nan],
    'year':[2000, 2001, 2002, 2001, 2002, 2099],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 99]
}

df1 = pd.DataFrame(
    data, 
    columns=['year', 'state', 'pop']
)

df1
```

```python
data = {
    'state': ['Ohio', 'Nevada', 'Utah', np.nan],
    'area':[100, 99, 88, 77]
}

df2 = pd.DataFrame(
    data, 
    columns=['state', 'area']
)

df2
```

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', np.nan],
    'year':[2020, 2021, 2022, 2021, 2022, 2099],
    'tem':[1, 2, 3, 4, 5, 6]
}

df3 = pd.DataFrame(
    data, 
    columns=['state', 'year', 'tem']
)

df3
```

```python
df4 = pd.DataFrame(np.arange(18).reshape(6, 3), columns=['tem', 'B', 'C'])

df4
```

## 横向拓展-df1横向merge拓展df2

```python
pd.merge(df1, df2, on=['state'], how='inner')   # warning! 空值会参与匹配
```

## 纵向拓展-df1纵向cocat拓展df2

```python
pd.concat([df1, df2, df3], axis=0, join='outer', ignore_index=True)  # ignore_index 会重置index
# 相同的列纵向拓展
# 不重叠的列，用空值填充增加的位置
# 可以合并多个df
```

## 默认按index横向拓展

```python
df3.join(df4, how='left', lsuffix='_j', rsuffix='_r') 
# 可拼接过个df
# 默认按index拼接
```

# 函数的应用和映射

```python
df5 = df3.join(df4, how='left', lsuffix='_j', rsuffix='_r') 

df5.head()
```

## 对一列进行map

```python
df5['str_num'] = df5['B'].map(lambda x: "'"+str(x))   # 在数据前面添加引号，在to_csv场景下输出不会被截断
df5
```

## 根据字典进行映射

```python
asist_map_dict={
    'Ohio': 'O',
    'Nevada': 'N'
}

df5['new_state'] = df5['state'].map(asist_map_dict)
df5
```
