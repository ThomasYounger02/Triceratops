# 示例

```python
# Data
x=[1, 2, 3, 4]
y=[1, 4, 9, 16]

# Plot
plt.plot(x, y, color='r', marker='o', linestyle='dashed', label='legend')  # 红色线条，圆圈标记，虚线连接

#  Decoration
plt.axis([0, 6, 0, 20])                                   # [x轴最小值, x轴最大值, y轴最小值, y轴最大值]
plt.xlabel('x_axis_label')                                # x轴标签
plt.ylabel('y_axis_label')                                # y轴标签
plt.title('Title')                                        # 标签
plt.legend()                                              # 图例

plt.annotate('comment', xy=(2,5), xytext=(2, 10),               # xy：箭头注释中箭头所在位置，xytext：注释文本所在位置
            arrowprops=dict(facecolor='black', shrink=0.01)     # arrowprops在xy和xytext之间绘制箭头, shrink表示注释点与注释文本之间的图标距离
            )                                              
# Show
plt.show()
```

```python
# Data
x = np.linspace(0, 10, 30)

# Plot
fig = plt.figure(figsize=(8, 6))            # 创建画布
plt.plot(x, np.sin(x), '--', label='sin')    # 虚线
plt.plot(x, np.cos(x), '-', label='cos')    # 实线

# Decoration
plt.xlim(0, 10)                         # x轴范围
plt.ylim(-1.5, 1.5)                     # y轴范围
plt.xlabel('x_lable')                   # x轴标签
plt.ylabel('y_lable')                   # y轴标签
plt.title('Title')                      # 标题
plt.grid()                              # 显示网格
plt.legend(loc='lower left', frameon=True, ncol=2)    # 左下角，显示边框，分成两列

# 辅助线
plt.axhline(y=0.8, ls='--', c='r')                               # 绘制y=0.8的水平参考线
plt.axvline(x=2.5, ls='--', c='r')

# 辅助区域
plt.axvspan(xmin=4, xmax=6, facecolor='r', alpha=0.3)            # 垂直x轴，参考区域
plt.axhspan(ymin=-0.2, ymax=0.2, facecolor='y', alpha=0.3);      # 垂直y轴，参考区域

# 注释文本
plt.text(3.2, 0, 'sin(x)', weight='bold', color='r')              # 注释文字
plt.annotate('maximum', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1),
             weight='bold',
             color='r',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'))   # 箭头标识

# Show
plt.show()
```

# 组件

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100, endpoint=True)    # 指定区间内均匀间隔，数量为20的数组向量
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(x-0.8)

plt.plot(x, y1, 
    color='g',       # c-青色；m-紫色；k-黑色
    linewidth=4, 
    linestyle='--',
    label=r'$y=sin(x)$'
    )
plt.plot(x, y2, 
    '*',
    markersize=8,
    markerfacecolor='r',
    markeredgecolor='k',
    label=r'$y=cos(x)$'
    )
plt.plot(x, y3,
    'k^-'   # color/marker/linestyle
    )

plt.legend(loc='best')
plt.show()
```

# 组件2

```python
x = np.arange(-5, 5, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)

# 点标记
for a, b in zip(x, y1):
    if abs(a-4)<1e-10:
        plt.text(a, b,             # 坐标
                (round(a,1), round(b,1)),   # 文本
                 ha='center',      # 水平对器
                 va='bottom', 
                 fontsize=10
                )
# # 做图
plt.plot(x, y1, 'r-', label=r'$y1=sin(x)$')
plt.plot(x, y2, 'b:', label=r'$y2=cos(x)$')
# 辅助线

# 组件控制
plt.xlim(-6, 6)
plt.ylim(-1.2, 1.2)
# plt.xticks(ticks=np.arange(-1.5*np.pi, 2*np.pi, 0.5*np.pi),   # 刻度范围
#           labels=['a', 'b', 'c', 'd', 'e', 'f', 'g'])         # 刻度标签
# plt.yticks(ticks=[-1, 0, 1])
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# 组件控制2
plt.grid(True, axis=u'y')        # 添加网格线, axis=默认both
plt.title('Lines-title')
plt.legend(loc='best')
plt.savefig('./result/demo-matplotlib-line.png', dpi=600)    # dpi-dots per inch,每英寸像素点
```

## 直方图

```python
mu = 100
sigma = 15
x = mu+sigma*np.random.randn(200)

num_bins = 25    # 划分组数

plt.figure(figsize=(16, 6), dpi=100)
n, bins, patches = plt.hist(x, num_bins,        #n-直方图数量，bins-分组区间，pathches-间隔信息
                           color='w',
                           edgecolor='k',
                           hatch=r'ooo',
                           density=True,        # 是否归一化处理数据
                           label = 'frq')
y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))


plt.plot(bins, y, '--', label='kde_f')

plt.xlabel('rate')
plt.ylabel('kde')
plt.title('$\mu=100$, $\sigma=15$')
plt.legend();
```

# 赛博朋克风mplcyberpunk

```python
# ! pip install mplcyberpunk 
import mplcyberpunk
```

```python
x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# 启用样式
plt.style.use('cyberpunk')
plt.figure(figsize=(16, 4))

# 折线图
plt.plot(x, y1, marker='o')
plt.plot(x, y2, marker='o', c='lime')

# 发光效果+渐变填充
mplcyberpunk.add_gradient_fill(
    alpha_gradientglow=0.5,
    gradient_start='zero')

# 坐标名称
plt.xlabel('X-axis')
plt.ylabel('Y-axis');
```
