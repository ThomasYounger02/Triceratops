# Functions

## plot_ts

```python
def plot_ts(series, title='fig_title', save=False, figsize=(16, 9)):
    """
    plot timeseries fig.
    
    Params:
        series - df or df.series, and index need to be a timeseries.
        title - fig title.

    Return:
        None
    """
    plt.figure(figsize=figsize)
    plt.plot(series)
    plt.title(title)
    plt.grid(True)
    
    if save:
        plt.savefig(f'{title}.jpg')    # 保存图片，跟在每个作图代码的最后，要在show之前.
    plt.show()                         # Q：为什么要加这个？作用是什么？
```

## plot_moving_average

```python
# 移动平均作图

def moving_average(series, n):
    """
    Calculate average of last n observations.
    """
    return np.average(series[-n: ])

def mean_absolute_percentage_error(y_true, y_pred):
    """
    Calculate mape of 2 series.
    """
    pe = (y_true - y_pred) / y_true
    mape = np.mean(np.abs(pe))*100
    return mape
```

```python
def plot_moving_average(series, window, plot_intervals=False, scale=1.96, plot_anomalies=False):
    """
    series - dataframe with timeseries
    window - rolling window size
    plot_intervals - show confidence intervals
    plot_anomalies - show anomalies
    """
    rolling_mean = series.rolling(window=window).mean()

    plt.figure(figsize=(15,5))
    plt.title("Moving average\\n window size = {}".format(window))
    plt.plot(rolling_mean, "g", label="Rolling mean trend")

    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * deviation)
        upper_bond = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bond, "r--", label="Upper Bond / Lower Bond")
        plt.plot(lower_bond, "r--")

    # Having the intervals, find abnormal values
    if plot_anomalies:
        anomalies = pd.DataFrame(index=series.index, columns=series.columns)
        anomalies[series<lower_bond] = series[series<lower_bond]
        anomalies[series>upper_bond] = series[series>upper_bond]
        plt.plot(anomalies, "ro", markersize=10)

    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="upper left")
    plt.grid(True)
```

```python
from sklearn.datasets import load_iris

data_iris = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
data_iris['sample'] = load_iris().target

# 列名中包含" (cm)"， 进行删除
for col in data_iris.columns:
    if ' (cm)' in col:
        new_col = col.replace(' (cm)', '')
        data_iris.rename(columns={col: new_col}, inplace=True)

data_iris.head()
```

## 构建-数据

`data_rts`

```python
# 加载-数据from seaborn
rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2022", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data_rts = data.rolling(7).mean()

data_rts.tail()
```

```python
# 下载-数据from github
# url = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/ads.csv"
# data_df = pd.read_csv(url)

data_ads = pd.read_csv('ads.csv')

data_ads['randn'] = np.random.randn(len(data_ads['Ads']))   # 生成一个新列，用于二元绘图
data_ads['date'] = pd.to_datetime(data_ads['Time'])         # 日期格式转换
data_ads['cut_1'] = data_ads['Ads'].map(lambda x: 'H' if x>=16*10000 else('M' if x>=8*10000 else 'L'))
data_ads['cut_2'] = np.where(data_ads['randn']>0,'P', 'NP')
data_ads['year'] = data_ads['date'].dt.year

data_ads.sample(5)
```

```python
# 数据准备2-设定日期为新的index
new_df = data_ads[['Ads', 'date']].set_index('date')

new_df.head()
```

## 散点图

```python
# 散点图，拆分维度组合后，进行散点图作图

# Q:不能在前面增加 > f, ax = plt.subplots(figsize=(8, 8))

# colors = ["black", "lightsteelblue", 'royalblue']
sns.relplot(
    data=data_ads,
    x="Ads", y="randn",                   # 两列数据
    row="year",                           # 按cut2拆分行
    col="cut_2",                          # 按cut1拆分列
    hue="cut_1",                          # 标识不同的点
#     palette=colors,                       # 对标hue的颜色数量
    kind="scatter",                       # 散点图
    height=5,
    aspect=.75, facet_kws=dict(sharex=True)
);
```

```python
# 按row*col组合后拆分子图，每个子图是一个x*y散点图,效果不如sns.relplot(kind=scatter)

g = sns.FacetGrid(data_ads,
row='cut_1',
col="cut_2",
hue="year",
margin_titles=False
)
g.map(sns.regplot,
"Ads", "randn",
data = data_ads,
fit_reg=False,
x_jitter=0.2);
```

## 组合作图

```python
# 分布+集中点：两侧kde用于分布查看，中间为交叉密度图，可用于判断聚合集中位置

g = sns.JointGrid("Ads", "randn", data_ads
# xlim=(-6,6),
# ylim=(-5,5)
)
g = g.plot_joint(sns.kdeplot, cmap="Blues", shade=True)
g = g.plot_marginals(sns.kdeplot, shade=True)
```
