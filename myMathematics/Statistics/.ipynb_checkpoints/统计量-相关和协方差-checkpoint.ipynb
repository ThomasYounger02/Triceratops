# 相关和协方差

## 1. Pearson相关方法和Spearman相关方法的比较

### 1.1. 什么是相关？

相关系数可度量两个变量倾向于同时变化的程度。该系数同时描述了关系的强度和方向。

**Pearson 积矩相关**

Pearson 相关评估两个连续变量之间的线性关系。当一个变量中的变化与另一个变量中的成比例变化相关时，这两个变量具有线性关系。例如，您可能使用 Pearson 相关来评估您生产设施温度的升高是否与巧克力涂层的厚度减少有关。

**Spearman 秩次相关**

Spearman 相关评估两个连续或顺序变量之间的单调关系。在单调关系中，变量倾向于同时变化，但不一定以恒定的速率变化。

Spearman 相关系数基于每个变量的秩值（而非原始数据）。Spearman 相关通常用于评估与顺序变量相关的关系。例如，您可能会使用 Spearman 相关来评估员工完成检验练习的顺序是否与他们工作的月数相关。

最好始终用散点图来检查变量之间的关系。相关系数仅度量线性 (Pearson) 或单调 (Spearman) 关系。也有可能存在其他关系。

### 1.2. 对 Pearson 和 Spearman 系数进行比较

Pearson 和 Spearman 相关系数的值介于 −1 和 +1 之间。当两个变量的上升量相同时，Pearson 相关系数为 +1。该关系会构成一条完美的直线。在本例中，Spearman 相关系数也是 +1。

![Pearson = +1，Spearman = +1](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/d6406b39-b23a-4ffa-bd76-de77160eaf8d/1.png)

Pearson = +1，Spearman = +1

如果关系是一个变量上升，其他变量也上升，但上升量不一致，则 Pearson 相关系数为小于 +1 的正数。在本例中，Spearman 系数仍等于 +1。

![Pearson = +0.851，Spearman = +1](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/80fe4fac-d7ab-4f3a-8297-64a18a116ffa/2.png)

Pearson = +0.851，Spearman = +1

当存在随机关系或者不存在关系时，这两个相关系数都接近零。

![Pearson = −0.093，Spearman = −0.093](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/cb0a2173-8c29-4f17-94b7-edafc7803637/3.png)

Pearson = −0.093，Spearman = −0.093

如果关系对应的是一条完美的下降关系直线，则这两个相关系数都为 −1。

![Pearson = −1，Spearman = −1](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/bc9f489c-19df-41d3-8227-c9200c3edaf9/4.png)

Pearson = −1，Spearman = −1

如果关系是一个变量下降，其他变量上升，但变化程度不一致，则 Pearson 相关系数为大于 −1 的负数。在本例中，Spearman 系数将仍等于 −1。

![Pearson = −0.799，Spearman = −1](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/0255bece-778b-4d90-826c-04628ff4d2b0/5.png)

Pearson = −0.799，Spearman = −1

相关值 −1 或 1 表明存在完全线性关系，例如，圆的半径和周长之间的关系。但是，在量化分析中，相关值的实际值小于完美关系的值。确定两个变量相关通常表示需要进行回归分析，从而尝试更详细地描述这种类型的关系。

### 1.3. 其他非线性关系

Pearson 相关系数只度量线性关系。Spearman 相关系数只度量单调关系。因此，即使相关系数为 0，也可能存在有意义的关系。检查散点图可确定关系的形式。

## 2. 什么是协方差？

协方差度量两个变量之间的线性关系。协方差与两个变量之间的相关相似，但是，它们在以下方面存在区别：

- 相关系数是标准化的。因此，完美的线性关系会生成系数 1。相关可度量两个变量之间线性关系的强度和方向。
- 协方差值不是标准化的。因此，协方差的范围可以为负无穷大到正无穷大。因此，完美线性关系的值取决于数据。因为数据不是标准化的，所以很难确定变量之间关系的强度。

可以使用协方差来了解变量之间关系的方向。正协方差值指示一个高于平均值的变量值与另一个高于平均值的变量值相关联，低于平均值的变量值也具有相似关联。负协方差值指示一个高于平均值的变量值与另一个低于平均值的变量值相关联。

协方差系数是协方差的函数。协方差系数等于协方差除以两个变量的标准差之积。因此，正协方差始终生成正相关，负协方差始终生成负相关。
